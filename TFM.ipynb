# Ejemplo 6

```python
m=2
K=2
r=4
activos = range(1,m+1)
fuentes = range(1,K+1)
restr = range(1,r+1)
Ns = [10,20]
Nacum = [0]
for i in range(K):
  Nacum = Nacum + [Nacum[i]+Ns[i]]
N = sum(Ns)

indices = []
for k in range(K):
  for i in range(Ns[k]):
    indices.append(k+1)

d = [1.1,1.0,0.3,0.2]
C = np.vstack((np.eye(m),-np.eye(m)))

beta = [1/3,2/3]
rho = 10
alfa = 0.2
eps = 0.01

coste_Pool = 0
x_Pool = np.zeros(m)
coste_MOWB = 0
x_MOWB = np.zeros(m)

repeticiones = 100
for prueba in range(repeticiones):
  xi = np.zeros((m,N))
  for n in range(Nacum[0], Nacum[1]):
    common = np.random.uniform(low=0, high=0.8)
    xi[0,n] = common + np.random.uniform(low=-0.2, high=0.1)
    xi[1,n] = common + np.random.uniform(low=-0.15, high=0.15)
  for n in range(Nacum[1], Nacum[2]):
    common = np.random.uniform(low=0.1, high=0.7)
    xi[0,n] = common + np.random.uniform(low=-0.4, high=0.4)
    xi[1,n] = common + np.random.uniform(low=-0.1, high=0.1)


  # Problema pool
  ampl.eval(r"""
  reset;                          # Para limpiar cualquier modelo previo
  param m;                        # Número de activos financieros
  param K;                        # Número de fuentes de datos
  param r;                        # Número de restricciones del conjunto de incertidumbre
  set ACTIVOS := 1..m;            # Índices de los activos financieros
  set FUENTES := 1..K;            # Índices de las fuentes de datos
  set RESTR := 1..r;              # Índices de las restricciones del conjunto de incertidumbre

  set ACUM := 0..K;
  param Nacum {ACUM};             # Número acumulado de observaciones de cada fuente, comenzando por 0

  param N;                        # Número total de observaciones
  set OBS := 1..N;                # Índices de las observaciones
  param xi {ACTIVOS, OBS};        # Conjunto de observaciones del rendimiento anual

  # Parámetros que definen el conjunto de incertidumbre
  param d {RESTR};
  param C {RESTR, ACTIVOS};

  param rho >= 0;                 # Codifica el riesgo
  param alfa >= 0;                # Significación
  param eps >= 0;                 # Radio de la bola de Wasserstein
  param beta {FUENTES} >= 0;      # Peso de cada fuente (basado en la confianza en este)

  # Variables de decisión
  var x {ACTIVOS} >= 0;           # Porcentaje a invertir en cada activo
  var tau;
  var lambda >=0;
  var s {OBS};
  var gamma1 {RESTR, OBS}>=0;
  var gamma2 {RESTR, OBS}>=0;

  # Problema de optimización
  minimize coste:
      sum{k in FUENTES} beta[k] * (lambda * eps + (sum{i in (Nacum[k-1]+1)..Nacum[k]} s[i])/(Nacum[k] - Nacum[k-1]));
  s.t. prob_simplex:
        sum{a in ACTIVOS} x[a] = 1;
      restric1 {n in OBS}:
        rho*tau - (sum{a in ACTIVOS} x[a]*xi[a,n]) + sum{b in RESTR} gamma1[b,n]*(d[b]- (sum{a in ACTIVOS} C[b,a]*xi[a,n])) <= s[n];
      restric2 {n in OBS}:
        rho* (1 - 1/alfa) *tau - (1 + rho/alfa) * (sum{a in ACTIVOS} x[a]*xi[a,n]) + sum{b in RESTR} gamma2[b,n]*(d[b]- (sum{a in ACTIVOS} C[b,a]*xi[a,n])) <= s[n];
      restric3 {a in ACTIVOS, n in OBS}:
        (sum{b in RESTR} C[b,a]*gamma1[b,n]) + x[a] <= lambda;
      restric4 {a in ACTIVOS, n in OBS}:
        (sum{b in RESTR} C[b,a]*gamma2[b,n]) + (1 + rho/alfa) * x[a] <= lambda;
  """)

  ampl.param["m"] = m
  ampl.param["K"] = K
  ampl.param["r"] = r
  ampl.param["Nacum"] = {a: b for a, b in enumerate(Nacum)}
  ampl.param["N"] = N
  ampl.param["d"] = {restr[i]: j for i, j in enumerate(d)}
  ampl.param["C"] = {
      (a, b): C[i][j]
      for i, a in enumerate(restr)
      for j, b in enumerate(activos)
  }
  ampl.param["xi"] = {
      (a, n): xi[i,j]
      for i, a in enumerate(activos)
      for j, n in enumerate(range(1,N+1))
  }
  ampl.param["rho"] = rho
  ampl.param["alfa"] = alfa
  ampl.param["eps"] = eps
  ampl.param["beta"] = {fuentes[i]: j for i, j in enumerate(beta)}

  ampl.solve(solver="highs") # Resolvemos con el solver "highs"
  coste_Pool += ampl.get_objective("coste").value()
  x_AMPL = ampl.get_variable("x") # Aquí el valor de "x" es convertido a un data frame de AMPL.
  x_df = x_AMPL.get_values().to_pandas() # Aquí el valor de "x" es convertido a un data frame de Pandas
  x = x_df['x.val']
  for i in range(m):
    x_Pool[i] += x[i+1]




  # Problema MoWB
  ampl.eval(r"""
  reset;                          # Para limpiar cualquier modelo previo
  param m;                        # Número de activos financieros
  param K;                        # Número de fuentes de datos
  param r;                        # Número de restricciones del conjunto de incertidumbre
  set ACTIVOS := 1..m;            # Índices de los activos financieros
  set FUENTES := 1..K;            # Índices de las fuentes de datos
  set RESTR := 1..r;              # Índices de las restricciones del conjunto de incertidumbre

  set ACUM := 0..K;
  param Nacum {ACUM};             # Número acumulado de observaciones de cada fuente, comenzando por 0

  param N;                        # Número total de observaciones
  set OBS := 1..N;                # Índices de las observaciones
  param INDICES {OBS};            # Indica de que fuente es cada observacion (ordenadas por fuente)
  param xi {ACTIVOS, OBS};        # Conjunto de observaciones del rendimiento anual

  # Parámetros que definen el conjunto de incertidumbre
  param d {RESTR};
  param C {RESTR, ACTIVOS};

  param rho >= 0;                  # Codifica el riesgo
  param alfa >= 0;                # Significación
  param eps {FUENTES} >= 0;       # Radio de la bola de Wasserstein
  param beta {FUENTES} >= 0;      # Peso de cada fuente (basado en la confianza en este)

  # Variables de decisión
  var x {ACTIVOS} >= 0;           # Porcentaje a invertir en cada activo
  var tau;
  var lambda {FUENTES} >=0;
  var s {OBS};
  var gamma1 {RESTR, OBS}>=0;
  var gamma2 {RESTR, OBS}>=0;

  # Problema de optimización
  minimize coste:
      sum{k in FUENTES} beta[k] * (lambda[k] * eps[k] + (sum{i in (Nacum[k-1]+1)..Nacum[k]} s[i])/(Nacum[k] - Nacum[k-1]));
  s.t. prob_simplex:
        sum{a in ACTIVOS} x[a] = 1;
      restric1 {n in OBS}:
        rho*tau - (sum{a in ACTIVOS} x[a]*xi[a,n]) + sum{b in RESTR} gamma1[b,n]*(d[b]- (sum{a in ACTIVOS} C[b,a]*xi[a,n])) <= s[n];
      restric2 {n in OBS}:
        rho* (1 - 1/alfa) *tau - (1 + rho/alfa) * (sum{a in ACTIVOS} x[a]*xi[a,n]) + sum{b in RESTR} gamma2[b,n]*(d[b]- (sum{a in ACTIVOS} C[b,a]*xi[a,n])) <= s[n];
      restric3 {a in ACTIVOS, n in OBS}:
        (sum{b in RESTR} C[b,a]*gamma1[b,n]) + x[a] <= lambda[(INDICES[n])];
      restric4 {a in ACTIVOS, n in OBS}:
        (sum{b in RESTR} C[b,a]*gamma2[b,n]) + (1 + rho/alfa) * x[a] <= lambda[(INDICES[n])];
  """)

  ampl.param["m"] = m
  ampl.param["K"] = K
  ampl.param["r"] = r
  ampl.param["Nacum"] = {a: b for a, b in enumerate(Nacum)}
  ampl.param["N"] = N
  ampl.param["INDICES"] = {a+1: b for a, b in enumerate(indices)}
  ampl.param["d"] = {restr[i]: j for i, j in enumerate(d)}
  ampl.param["C"] = {
      (a, b): C[i][j]
      for i, a in enumerate(restr)
      for j, b in enumerate(activos)
  }
  ampl.param["xi"] = {
      (a, n): xi[i,j]
      for i, a in enumerate(activos)
      for j, n in enumerate(range(1,N+1))
  }
  ampl.param["rho"] = rho
  ampl.param["alfa"] = alfa
  eps2 = [eps,eps]
  ampl.param["eps"] = {fuentes[i]: j for i, j in enumerate(eps2)}
  ampl.param["beta"] = {fuentes[i]: j for i, j in enumerate(beta)}

  ampl.solve(solver="highs") # Resolvemos con el solver "highs"
  coste_MOWB += ampl.get_objective("coste").value()
  x_AMPL = ampl.get_variable("x") # Aquí el valor de "x" es convertido a un data frame de AMPL.
  x_df = x_AMPL.get_values().to_pandas() # Aquí el valor de "x" es convertido a un data frame de Pandas
  x = x_df['x.val']
  for i in range(m):
    x_MOWB[i] += x[i+1]

print("Coste medio Pool: ")
print(coste_Pool/repeticiones)
print("Solución media Pool: ")
print(x_Pool/repeticiones)
print("Coste medio MoWB: ")
print(coste_MOWB/repeticiones)
print("Solución media MoWB: ")
print(x_MOWB/repeticiones)
```

```python
m = 4
r = 1
K = 2
activos = range(1,m+1)
restr = range(1,r+1)
fuentes = range(1,K+1)
Ns = [10,20]
Nacum = [0]
for i in range(K):
  Nacum = Nacum + [Nacum[i]+Ns[i]]
N = sum(Ns)

indices = []
for k in range(K):
  for i in range(Ns[k]):
    indices.append(k+1)

d = [0]
C = np.zeros((1,m))

beta = [1/3,2/3]
rho = 10
alfa = 0.2
eps = 0.025


xi = np.zeros((K,m,Ns[0]+Ns[1]))
for i in range(Ns[0]):
  common = np.random.normal(loc=0, scale=2)
  for a in range(2):
    xi[0,a,i] = common + np.random.normal(loc=5, scale=1)
  for a in range(2,4):
    xi[0,a,i] = common + np.random.normal(loc=-5, scale=1)
for i in range(Ns[1]):
  common = np.random.normal(loc=0, scale=2)
  for a in range(2):
    xi[1,a,i] = common + np.random.normal(loc=4, scale=1)
  for a in range(2,m):
    xi[1,a,i] = common + np.random.normal(loc=-3, scale=1)

xijunto = np.zeros((m,Ns[0]+Ns[1]))
for i in range(Ns[0]):
  xijunto[0,i] = xi[0,0,i]
  xijunto[1,i] = xi[0,1,i]
for i in range(Ns[1]):
  xijunto[0,i+Ns[0]] = xi[1,0,i]
  xijunto[1,i+Ns[0]] = xi[1,1,i]


L = 300
h = [(1/(i+3))**(1/2) for i in range(1,L+1)]

xl = np.ones(m)*(1/m)
taul = 0

def dl(x, tau, xi, m, rho, alfa):
  prod_esc = 0
  for i in range(m):
    prod_esc = prod_esc + x[i]*xi[i]
  l1 = -prod_esc + rho *tau
  l2 = - (1 + rho/alfa)*prod_esc + rho*(1 - 1/alfa)*tau
  l = max(l1,l2)
  if l == l1:
    sol = np.concatenate((-xi,np.array([rho])))
  else:
    sol = np.concatenate((-(1 + rho/alfa)*xi,np.array([rho*(1 - 1/alfa)])))
  return sol

for l in range(L):
  fk = np.zeros(K)
  vk = np.zeros((K,m+1))
  for k in range(K):
    ampl.eval(r"""
    reset;                          # Para limpiar cualquier modelo previo
    param m;                        # Número de activos financieros
    set ACTIVOS := 1..m;            # Índices de los activos financieros
    param r;                        # Número de restricciones del conjunto de incertidumbre
    set RESTR := 1..r;              # Índices de las restricciones del conjunto de incertidumbre
    param Nk;                       # Número total de observaciones
    set OBS := 1..Nk;               # Índices de las observaciones
    param xi {ACTIVOS, OBS};        # Conjunto de observaciones del rendimiento anual

    # Parámetros que definen el conjunto de incertidumbre
    param d {RESTR};
    param C {RESTR, ACTIVOS};

    param rho >= 0;                 # Codifica el riesgo
    param alfa >= 0;                # Significación
    param eps >= 0;                 # Radio de la bola de Wasserstein

    param x {ACTIVOS} >= 0;        # Porcentaje a invertir en cada activo
    param tau;

    # Variables de decisión
    var alfa1 {OBS} >= 0;
    var alfa2 {OBS} >= 0;
    var q1 {ACTIVOS, OBS};
    var q2 {ACTIVOS, OBS};

    # Problema de optimización
    maximize coste:
        (1/Nk) * sum{i in OBS} ( (alfa1[i]*rho*tau - (sum{a in ACTIVOS} x[a]*(alfa1[i]*xi[a,i]- q1[a,i]))) +
          (alfa2[i]*rho*(1 - 1/alfa)*tau - (1 + rho/alfa)*(sum{a in ACTIVOS} x[a]*(alfa2[i]*xi[a,i] - q2[a,i]))) );
    s.t. restric1:
          sum{i in OBS} ((sum{a in ACTIVOS} abs(q1[a,i])) + (sum{a in ACTIVOS} abs(q2[a,i]))) <= eps*Nk;
        restric2 {i in OBS}:
          alfa1[i] + alfa2[i] = 1;
        restric3 {i in OBS, b in RESTR}:
          alfa1[i]*((sum{a in ACTIVOS} C[b,a]*xi[a,i])- d[b]) <= (sum{a in ACTIVOS} C[b,a]*q1[a,i]);
        restric4 {i in OBS, b in RESTR}:
          alfa2[i]*((sum{a in ACTIVOS} C[b,a]*xi[a,i])- d[b]) <= (sum{a in ACTIVOS} C[b,a]*q2[a,i]);
    """)
    ampl.param["m"] = m
    ampl.param["r"] = r
    ampl.param["Nk"] = Ns[k]
    ampl.param["d"] = {restr[i]: j for i, j in enumerate(d)}
    ampl.param["C"] = {
        (a, b): C[i][j]
        for i, a in enumerate(restr)
        for j, b in enumerate(activos)
    }
    ampl.param["xi"] = {
        (a, n): xi[k,i,j]
        for i, a in enumerate(activos)
        for j, n in enumerate(range(1,Ns[k]+1))
    }
    ampl.param["rho"] = rho
    ampl.param["alfa"] = alfa
    ampl.param["eps"] = eps

    ampl.param["x"] = {a: xl[i] for i, a in enumerate(activos)}
    ampl.param["tau"] = taul

    ampl.solve(solver="highs") # Resolvemos con el solver "highs"
    fk[k] = ampl.get_objective("coste").value()
    alfa1_AMPL = ampl.get_variable("alfa1") # Aquí el valor de "x" es convertido a un data frame de AMPL.
    alfa1_df = alfa1_AMPL.get_values().to_pandas().round(5) # Aquí el valor de "x" es convertido a un data frame de Pandas
    alfa1 = alfa1_df['alfa1.val']
    alfa2_AMPL = ampl.get_variable("alfa2") # Aquí el valor de "x" es convertido a un data frame de AMPL.
    alfa2_df = alfa2_AMPL.get_values().to_pandas().round(5) # Aquí el valor de "x" es convertido a un data frame de Pandas
    alfa2 = alfa2_df['alfa2.val']
    q1_AMPL = ampl.get_variable("q1") # Aquí el valor de "x" es convertido a un data frame de AMPL.
    q1_df = q1_AMPL.get_values().to_pandas().round(5) # Aquí el valor de "x" es convertido a un data frame de Pandas
    q1 = q1_df['q1.val']
    q2_AMPL = ampl.get_variable("q2") # Aquí el valor de "x" es convertido a un data frame de AMPL.
    q2_df = q2_AMPL.get_values().to_pandas().round(5) # Aquí el valor de "x" es convertido a un data frame de Pandas
    q2 = q2_df['q2.val']
    xi_opt1 = xi[k,:,:]
    for i in range(Ns[k]):
      if alfa1[i+1] > 0:
        for j in range(m):
          xi_opt1[j,i] -= q1[j+1,i+1]/alfa1[i+1]
    xi_opt2 = xi[k,:,:]
    for i in range(Ns[k]):
      if alfa2[i+1] > 0:
        for j in range(m):
          xi_opt2[j,i] -= q2[j+1,i+1]/alfa2[i+1]
    for i in range(Ns[k]):
      vk[k,:] += dl(xl, taul, xi_opt1[:,i], m, rho, alfa) * alfa1[i+1]
      vk[k,:] += dl(xl, taul, xi_opt2[:,i], m, rho, alfa) * alfa2[i+1]
    vk[k,:] = vk[k,:]/Ns[k]
  fl = 0
  for k in range(K):
    fl += beta[k]*fk[k]
  if l==0:
    f_mejor = fl
    x_mejor = xl
  else:
    f_mejor = min(fl, f_mejor)
    if f_mejor==fl:
      x_mejor = xl

  vl = np.zeros(m+1)
  for k in range(K):
    vl += beta[k]*vk[k,:]
  xl = xl - h[l]*vl[0:m]
  ampl.eval(r"""
  reset;                          # Para limpiar cualquier modelo previo
  param m;                        # Número de activos financieros
  set ACTIVOS := 1..m;            # Índices de los activos financieros
  param x {ACTIVOS};        # Porcentaje a invertir en cada activo

  # Variables de decisión
  var xnew {ACTIVOS} >= 0;

  # Problema de optimización
  minimize distancia:
        sum{a in ACTIVOS} (xnew[a] - x[a])*(xnew[a] - x[a]);
  s.t. restric:
        sum{a in ACTIVOS} xnew[a] = 1;
  """)
  ampl.param["m"] = m
  ampl.param["x"] = {a: xl[i] for i, a in enumerate(activos)}
  ampl.solve(solver="highs") # Resolvemos con el solver "highs"
  xnew_AMPL = ampl.get_variable("xnew") # Aquí el valor de "x" es convertido a un data frame de AMPL.
  xnew_df = xnew_AMPL.get_values().to_pandas().round(5) # Aquí el valor de "x" es convertido a un data frame de Pandas
  xnew = xnew_df['xnew.val']
  for a in range(m):
    xl[a] = xnew[a+1]
  taul = taul - h[l]*vl[m]
print("solución Algoritmo 1: ", x_mejor)



# Problema MoWB
ampl.eval(r"""
reset;                          # Para limpiar cualquier modelo previo
param m;                        # Número de activos financieros
param K;                        # Número de fuentes de datos
param r;                        # Número de restricciones del conjunto de incertidumbre
set ACTIVOS := 1..m;            # Índices de los activos financieros
set FUENTES := 1..K;            # Índices de las fuentes de datos
set RESTR := 1..r;              # Índices de las restricciones del conjunto de incertidumbre

set ACUM := 0..K;
param Nacum {ACUM};             # Número acumulado de observaciones de cada fuente, comenzando por 0

param N;                        # Número total de observaciones
set OBS := 1..N;                # Índices de las observaciones
param xi {ACTIVOS, OBS};        # Conjunto de observaciones del rendimiento anual

# Parámetros que definen el conjunto de incertidumbre
param d {RESTR};
param C {RESTR, ACTIVOS};

param rho >= 0;                 # Codifica el riesgo
param alfa >= 0;                # Significación
param eps >= 0;                 # Radio de la bola de Wasserstein
param beta {FUENTES} >= 0;      # Peso de cada fuente (basado en la confianza en este)

# Variables de decisión
var x {ACTIVOS} >= 0;           # Porcentaje a invertir en cada activo
var tau;
var lambda >=0;
var s {OBS};
var gamma1 {RESTR, OBS}>=0;
var gamma2 {RESTR, OBS}>=0;

# Problema de optimización
minimize coste:
    sum{k in FUENTES} beta[k] * (lambda * eps + (sum{i in (Nacum[k-1]+1)..Nacum[k]} s[i])/(Nacum[k] - Nacum[k-1]));
s.t. prob_simplex:
      sum{a in ACTIVOS} x[a] = 1;
    restric1 {n in OBS}:
      rho*tau - (sum{a in ACTIVOS} x[a]*xi[a,n]) + sum{b in RESTR} gamma1[b,n]*(d[b]- (sum{a in ACTIVOS} C[b,a]*xi[a,n])) <= s[n];
    restric2 {n in OBS}:
      rho* (1 - 1/alfa) *tau - (1 + rho/alfa) * (sum{a in ACTIVOS} x[a]*xi[a,n]) + sum{b in RESTR} gamma2[b,n]*(d[b]- (sum{a in ACTIVOS} C[b,a]*xi[a,n])) <= s[n];
    restric3 {a in ACTIVOS, n in OBS}:
      (sum{b in RESTR} C[b,a]*gamma1[b,n]) + x[a] <= lambda;
    restric4 {a in ACTIVOS, n in OBS}:
      (sum{b in RESTR} C[b,a]*gamma2[b,n]) + (1 + rho/alfa) * x[a] <= lambda;
""")

ampl.param["m"] = m
ampl.param["K"] = K
ampl.param["r"] = r
ampl.param["Nacum"] = {a: b for a, b in enumerate(Nacum)}
ampl.param["N"] = N
ampl.param["d"] = {restr[i]: j for i, j in enumerate(d)}
ampl.param["C"] = {
    (a, b): C[i][j]
    for i, a in enumerate(restr)
    for j, b in enumerate(activos)
}
ampl.param["xi"] = {
    (a, n): xijunto[i,j]
    for i, a in enumerate(activos)
    for j, n in enumerate(range(1,N+1))
}
ampl.param["rho"] = rho
ampl.param["alfa"] = alfa
ampl.param["eps"] = eps
ampl.param["beta"] = {fuentes[i]: j for i, j in enumerate(beta)}

ampl.solve(solver="highs") # Resolvemos con el solver "highs"
x_AMPL = ampl.get_variable("x") # Aquí el valor de "x" es convertido a un data frame de AMPL.
x_df = x_AMPL.get_values().to_pandas() # Aquí el valor de "x" es convertido a un data frame de Pandas
x_df = x_df.round(5)
x = x_df['x.val']
x_sol = np.zeros(m)
for a in range(m):
  x_sol[a] = x[a+1]
print("solución MoWB: ", x_sol)
```
