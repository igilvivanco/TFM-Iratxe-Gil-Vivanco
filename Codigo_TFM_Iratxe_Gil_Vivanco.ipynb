# Ejemplo 6

```python
# Definimos los parámetros
m=2 # Número de activos financieros
K=2 # Número de fuentes de datos
r=4 # Número de restricciones del conjunto del conjunto de incertidumbre
activos = range(1,m+1) # Índices de los activos financieros
fuentes = range(1,K+1) # Índices de las fuentes de datos
restr = range(1,r+1) # Índices de las restricciones del conjunto de incertidumbre

Ns = [5,5] # Número de observaciones de cada fuente
Nacum = [0] # Número acumulado de observaciones de cada fuente, comenzando por 0
for i in range(K):
  Nacum = Nacum + [Nacum[i]+Ns[i]]
N = sum(Ns) # Número de observaciones totales

indices = [] # Indica de que fuente es cada observacion (ordenadas por fuente)
for k in range(K):
  for i in range(Ns[k]):
    indices.append(k+1)

# Parámetros que definen el conjunto de incertidumbre
d = [2,2,1,1]
C = np.vstack((np.eye(m),-np.eye(m)))

beta = [1/2,1/2] # Pesos de cada fuente de datos
rho = 10 # Parámetro que codifica el riesgo que queremos asumir
alfa = 0.2 # Significación del Valor en Riesgo Condicional
eps = 0.005 # Radio de las bolas de Wassertein

# Costes y soluciones medias
coste_Pool = 0
x_Pool = np.zeros(m)
coste_MOWB = 0
x_MOWB = np.zeros(m)

repeticiones = 100 # Número de pruebas

# Realizamos las pruebas
for prueba in range(repeticiones):

  # Definimos la muestra
  xi = np.zeros((m,N))
  for n in range(Nacum[0], Nacum[1]):
    common = np.random.uniform(low=0, high=1.3)
    xi[0,n] = common + np.random.uniform(low=-0.2, high=0.6)
    xi[1,n] = common + np.random.uniform(low=-0.5, high=0.5)
  for n in range(Nacum[1], Nacum[2]):
    common = np.random.uniform(low=-0.2, high=1.0)
    xi[0,n] = common + np.random.uniform(low=-0.4, high=0.1)
    xi[1,n] = common + np.random.uniform(low=-0.3, high=0.3)

  # Problema Pool-DRO
  ampl.eval(r"""
  reset;                          # Para limpiar cualquier modelo previo

  # Definimos los parámetros
  param m;                        # Número de activos financieros
  param K;                        # Número de fuentes de datos
  param r;                        # Número de restricciones del conjunto de incertidumbre
  set ACTIVOS := 1..m;            # Índices de los activos financieros
  set FUENTES := 1..K;            # Índices de las fuentes de datos
  set RESTR := 1..r;              # Índices de las restricciones del conjunto de incertidumbre

  set ACUM := 0..K;
  param Nacum {ACUM};             # Número acumulado de observaciones de cada fuente, comenzando por 0

  param N;                        # Número total de observaciones
  set OBS := 1..N;                # Índices de las observaciones
  param xi {ACTIVOS, OBS};        # Conjunto de observaciones de los rendimientos anuales

  # Parámetros que definen el conjunto de incertidumbre
  param d {RESTR};
  param C {RESTR, ACTIVOS};

  param rho >= 0;                 # Codifica el riesgo
  param alfa >= 0;                # Significación del Valor en Riesgo Condicional
  param eps >= 0;                 # Radio de la bola de Wasserstein
  param beta {FUENTES} >= 0;      # Peso de cada fuente (basado en la confianza en este)

  # Definimos las variables de decisión
  var x {ACTIVOS} >= 0;           # Porcentaje a invertir en cada activo
  var tau;
  var lambda >=0;
  var s {OBS};
  var y1 {RESTR, OBS}>=0;
  var y2 {RESTR, OBS}>=0;

  # Definimos el problema de optimización
  minimize coste:
      sum{k in FUENTES} beta[k] * (lambda * eps + (sum{i in (Nacum[k-1]+1)..Nacum[k]} s[i])/(Nacum[k] - Nacum[k-1]));
  s.t. prob_simplex:
        sum{a in ACTIVOS} x[a] = 1;
      restric1 {i in OBS}:
        rho*tau - (sum{a in ACTIVOS} x[a]*xi[a,i]) + sum{b in RESTR} y1[b,i]*(d[b]- (sum{a in ACTIVOS} C[b,a]*xi[a,i])) <= s[i];
      restric2 {i in OBS}:
        rho* (1 - 1/alfa) *tau - (1 + rho/alfa) * (sum{a in ACTIVOS} x[a]*xi[a,i]) + sum{b in RESTR} y2[b,i]*(d[b]- (sum{a in ACTIVOS} C[b,a]*xi[a,i])) <= s[i];
      restric3 {a in ACTIVOS, i in OBS}:
        (sum{b in RESTR} C[b,a]*y1[b,i]) + x[a] <= lambda;
      restric4 {a in ACTIVOS, i in OBS}:
        (sum{b in RESTR} C[b,a]*y2[b,i]) + (1 + rho/alfa) * x[a] <= lambda;
  """)

  # Definimos los datos en AMPL
  ampl.param["m"] = m
  ampl.param["K"] = K
  ampl.param["r"] = r
  ampl.param["Nacum"] = {a: b for a, b in enumerate(Nacum)}
  ampl.param["N"] = N
  ampl.param["d"] = {restr[i]: j for i, j in enumerate(d)}
  ampl.param["C"] = {
      (a, b): C[i][j]
      for i, a in enumerate(restr)
      for j, b in enumerate(activos)
  }
  ampl.param["xi"] = {
      (a, n): xi[i,j]
      for i, a in enumerate(activos)
      for j, n in enumerate(range(1,N+1))
  }
  ampl.param["rho"] = rho
  ampl.param["alfa"] = alfa
  ampl.param["eps"] = eps
  ampl.param["beta"] = {fuentes[i]: j for i, j in enumerate(beta)}

  ampl.set_option("solver", "highs") # Seleccionamos el solver "highs"
  ampl.set_option("solver_msg", 0) # Silenciamos el mensaje de retorno de AMPL
  ampl.solve() # Resolvemos el problema
  coste_Pool += ampl.get_objective("coste").value() # Obtenemos el valor objetivo y lo sumamos a coste_Pool
  x_AMPL = ampl.get_variable("x") # Convertimos el valor de "x" a un data frame de AMPL
  x_df = x_AMPL.get_values().to_pandas() # Convertimos el valor de "x" a un data frame de Pandas
  # Obtenemos el valor de "x" en la solución y lo sumamos a x_Pool
  x = x_df['x.val']
  for i in range(m):
    x_Pool[i] += x[i+1]


  # Problema MoWB-DRO
  ampl.eval(r"""
  reset;                          # Para limpiar cualquier modelo previo

  # Definimos los parámetros
  param m;                        # Número de activos financieros
  param K;                        # Número de fuentes de datos
  param r;                        # Número de restricciones del conjunto de incertidumbre
  set ACTIVOS := 1..m;            # Índices de los activos financieros
  set FUENTES := 1..K;            # Índices de las fuentes de datos
  set RESTR := 1..r;              # Índices de las restricciones del conjunto de incertidumbre

  set ACUM := 0..K;
  param Nacum {ACUM};             # Número acumulado de observaciones de cada fuente, comenzando por 0

  param N;                        # Número total de observaciones
  set OBS := 1..N;                # Índices de las observaciones
  param INDICES {OBS};            # Indica de que fuente es cada observacion (ordenadas por fuente)
  param xi {ACTIVOS, OBS};        # Conjunto de observaciones de los rendimientos anuales

  # Parámetros que definen el conjunto de incertidumbre
  param d {RESTR};
  param C {RESTR, ACTIVOS};

  param rho >= 0;                  # Codifica el riesgo
  param alfa >= 0;                 # Significación
  param eps {FUENTES} >= 0;        # Radio de las bolas de Wasserstein
  param beta {FUENTES} >= 0;       # Peso de cada fuente (basado en la confianza en este)

  # Definimos las variables de decisión
  var x {ACTIVOS} >= 0;            # Porcentaje a invertir en cada activo
  var tau;
  var lambda {FUENTES} >=0;
  var s {OBS};
  var y1 {RESTR, OBS}>=0;
  var y2 {RESTR, OBS}>=0;

  # Definimos el problema de optimización
  minimize coste:
      sum{k in FUENTES} beta[k] * (lambda[k] * eps[k] + (sum{i in (Nacum[k-1]+1)..Nacum[k]} s[i])/(Nacum[k] - Nacum[k-1]));
  s.t. prob_simplex:
        sum{a in ACTIVOS} x[a] = 1;
      restric1 {i in OBS}:
        rho*tau - (sum{a in ACTIVOS} x[a]*xi[a,i]) + sum{b in RESTR} y1[b,i]*(d[b]- (sum{a in ACTIVOS} C[b,a]*xi[a,i])) <= s[i];
      restric2 {i in OBS}:
        rho* (1 - 1/alfa) *tau - (1 + rho/alfa) * (sum{a in ACTIVOS} x[a]*xi[a,i]) + sum{b in RESTR} y2[b,i]*(d[b]- (sum{a in ACTIVOS} C[b,a]*xi[a,i])) <= s[i];
      restric3 {a in ACTIVOS, i in OBS}:
        (sum{b in RESTR} C[b,a]*y1[b,i]) + x[a] <= lambda[(INDICES[i])];
      restric4 {a in ACTIVOS, i in OBS}:
        (sum{b in RESTR} C[b,a]*y2[b,i]) + (1 + rho/alfa) * x[a] <= lambda[(INDICES[i])];
  """)

  # Definimos los datos en AMPL
  ampl.param["m"] = m
  ampl.param["K"] = K
  ampl.param["r"] = r
  ampl.param["Nacum"] = {a: b for a, b in enumerate(Nacum)}
  ampl.param["N"] = N
  ampl.param["INDICES"] = {a+1: b for a, b in enumerate(indices)}
  ampl.param["d"] = {restr[i]: j for i, j in enumerate(d)}
  ampl.param["C"] = {
      (a, b): C[i][j]
      for i, a in enumerate(restr)
      for j, b in enumerate(activos)
  }
  ampl.param["xi"] = {
      (a, n): xi[i,j]
      for i, a in enumerate(activos)
      for j, n in enumerate(range(1,N+1))
  }
  ampl.param["rho"] = rho
  ampl.param["alfa"] = alfa
  eps2 = [eps,eps]
  ampl.param["eps"] = {fuentes[i]: j for i, j in enumerate(eps2)}
  ampl.param["beta"] = {fuentes[i]: j for i, j in enumerate(beta)}


  ampl.set_option("solver", "highs") # Seleccionamos el solver "highs"
  ampl.set_option("solver_msg", 0) # Silenciamos el mensaje de retorno de AMPL
  ampl.solve() # Resolvemos el problema
  coste_MOWB += ampl.get_objective("coste").value() # Obtenemos el valor objetivo y lo sumamos a coste_MOWB
  x_AMPL = ampl.get_variable("x") # Convertimos el valor de "x" a un data frame de AMPL
  x_df = x_AMPL.get_values().to_pandas() # Convertimos el valor de "x" a un data frame de Pandas
  # Obtenemos el valor de "x" en la solución y lo sumamos a x_MOWB
  x = x_df['x.val']
  for i in range(m):
    x_MOWB[i] += x[i+1]

# Imprimimos, para cada método, la media de los valores objetivos y la media de las soluciones obtenidas
print(f"Coste medio Pool: {coste_Pool/repeticiones:.4f}")
print(f"Solución media Pool: {(x_Pool/repeticiones).round(4)}")
print(f"Coste medio MoWB: {coste_MOWB/repeticiones:.4f}")
print(f"Solución media MoWB: {(x_MOWB/repeticiones).round(4)}")
```

```python
# Función que evalua l
def l(x, tau, xi, m, rho, alfa):
  prod_esc = 0
  for i in range(m):
    prod_esc = prod_esc + x[i+1]*xi[i]
  return max(-prod_esc + rho *tau, - (1 + rho/alfa)*prod_esc + rho*(1 - 1/alfa)*tau)

# Función que devuelve la pérdida fuera de muestra aproximada
def perdida(x, tau, m, rho, alfa):
  npuntos = 750
  suma = 0
  for n in range(npuntos):
    xi = np.zeros(m)
    common = np.random.normal(loc=0, scale=0.2)
    for i in range(2):
      xi[i] = common + np.random.normal(loc=0.3, scale=0.1)
    for i in range(2,4):
      xi[i] = common + np.random.normal(loc=0.2, scale=0.1)
    for i in range(4,m):
      xi[i] = common + np.random.normal(loc=-0.4, scale=0.1)
    suma = suma + l(x,tau,xi,m,rho,alfa)
  return suma/npuntos

# Definimos los parámetros
m=6 # Número de activos financieros
K=2 # Número de fuentes de datos
r=1 # Número de restricciones del conjunto del conjunto de incertidumbre
activos = range(1,m+1) # Índices de los activos financieros
fuentes = range(1,K+1) # Índices de las fuentes de datos
restr = range(1,r+1) # Índices de las restricciones del conjunto de incertidumbre

Ns = [10,20] # Número de observaciones de cada fuente
Nacum = [0] # Número acumulado de observaciones de cada fuente, comenzando por 0
for i in range(K):
  Nacum = Nacum + [Nacum[i]+Ns[i]]
N = sum(Ns) # Número de observaciones totales

indices = [] # Indica de que fuente es cada observacion (ordenadas por fuente)
for k in range(K):
  for i in range(Ns[k]):
    indices.append(k+1)

# Parámetros que definen el conjunto de incertidumbre
d = [0]
C = np.zeros((1,m))

beta = [1/3,2/3] # Pesos de cada fuente de datos
rho = 10 # Parámetro que codifica el riesgo que queremos asumir
alfa = 0.2 # Significación del Valor en Riesgo Condicional

posibles_eps = [0.1,0.3,0.6] # Radios de las bolas de Wassertein que consideraremos
n_eps = len(posibles_eps) # Número de radios a utilizar
parejas_eps = [] # Parejas de radios a utilizar en MoWB-DRO
for i_eps in range(n_eps):
  for j_eps in range(n_eps):
    if i_eps != j_eps:
      parejas_eps.append([posibles_eps[i_eps],posibles_eps[j_eps]])
n_parejas_eps = len(parejas_eps) # Número de parejas de radios a utilizar en MoWB-DRO

repeticiones = 100 # Número de pruebas

perdidas_pool = np.zeros((n_eps,repeticiones)) # Pérdidas fuera de muestra de Pool-DRO
certificados_pool = np.zeros((n_eps,repeticiones)) # Certificados de Pool-DRO
fiabilidad_pool = np.zeros(n_eps) # Fiabilidad aproximada de Pool-DRO
perdidas_mowb = np.zeros((n_parejas_eps,repeticiones)) # Pérdidas fuera de muestra de MoWB-DRO
certificados_mowb = np.zeros((n_parejas_eps,repeticiones)) # Certificados de MoWB-DRO
fiabilidad_mowb = np.zeros(n_parejas_eps) # Fiabilidad aproximada de MoWB-DRO
perdidas_saa = np.zeros(repeticiones) # Pérdidas fuera de muestra de SAA
certificados_saa = np.zeros(repeticiones) # Certificados de SAA
fiabilidad_saa = 0 # Fiabilidad aproximada de muestra de SAA

# Realizamos las pruebas
for prueba in range(repeticiones):

  # Definimos la muestra
  xi = np.zeros((m,N))
  for n in range(Nacum[0], Nacum[1]):
    common = np.random.normal(loc=0, scale=0.2)
    for i in range(2):
      xi[i,n] = common + np.random.normal(loc=0.4, scale=0.1)
    for i in range(2,4):
      xi[i,n] = common + np.random.normal(loc=0.0, scale=0.1)
    for i in range(4,m):
      xi[i,n] = common + np.random.normal(loc=-0.4, scale=0.1)
  for n in range(Nacum[1], Nacum[2]):
    common = np.random.normal(loc=0, scale=0.05)
    for i in range(4):
      xi[i,n] = common + np.random.normal(loc=0.4, scale=0.1)
    for i in range(4,m):
      xi[i,n] = common + np.random.normal(loc=-0.4, scale=0.1)

  # Utilizamos Pool-DRO para cada radio
  for i_eps in range(n_eps):

    # Problema Pool-DRO
    ampl.eval(r"""
    reset;                          # Para limpiar cualquier modelo previo

    # Definimos los parámetros
    param m;                        # Número de activos financieros
    param K;                        # Número de fuentes de datos
    param r;                        # Número de restricciones del conjunto de incertidumbre
    set ACTIVOS := 1..m;            # Índices de los activos financieros
    set FUENTES := 1..K;            # Índices de las fuentes de datos
    set RESTR := 1..r;              # Índices de las restricciones del conjunto de incertidumbre

    set ACUM := 0..K;
    param Nacum {ACUM};             # Número acumulado de observaciones de cada fuente, comenzando por 0

    param N;                        # Número total de observaciones
    set OBS := 1..N;                # Índices de las observaciones
    param xi {ACTIVOS, OBS};        # Conjunto de observaciones de los rendimientos anuales

    # Parámetros que definen el conjunto de incertidumbre
    param d {RESTR};
    param C {RESTR, ACTIVOS};

    param rho >= 0;                 # Codifica el riesgo
    param alfa >= 0;                # Significación
    param eps >= 0;                 # Radio de la bola de Wasserstein
    param beta {FUENTES} >= 0;      # Peso de cada fuente (basado en la confianza en este)

    # Definimos las variables de decisión
    var x {ACTIVOS} >= 0;           # Porcentaje a invertir en cada activo
    var tau;
    var lambda >=0;
    var s {OBS};
    var y1 {RESTR, OBS}>=0;
    var y2 {RESTR, OBS}>=0;

    # Definimos el problema de optimización
    minimize coste:
        sum{k in FUENTES} beta[k] * (lambda * eps + (sum{i in (Nacum[k-1]+1)..Nacum[k]} s[i])/(Nacum[k] - Nacum[k-1]));
    s.t. prob_simplex:
          sum{a in ACTIVOS} x[a] = 1;
        restric1 {i in OBS}:
          rho*tau - (sum{a in ACTIVOS} x[a]*xi[a,i]) + sum{b in RESTR} y1[b,i]*(d[b]- (sum{a in ACTIVOS} C[b,a]*xi[a,i])) <= s[i];
        restric2 {i in OBS}:
          rho* (1 - 1/alfa) *tau - (1 + rho/alfa) * (sum{a in ACTIVOS} x[a]*xi[a,i]) + sum{b in RESTR} y2[b,i]*(d[b]- (sum{a in ACTIVOS} C[b,a]*xi[a,i])) <= s[i];
        restric3 {a in ACTIVOS, i in OBS}:
          (sum{b in RESTR} C[b,a]*y1[b,i]) + x[a] <= lambda;
        restric4 {a in ACTIVOS, i in OBS}:
          (sum{b in RESTR} C[b,a]*y2[b,i]) + (1 + rho/alfa) * x[a] <= lambda;
    """)

    # Definimos los datos en AMPL
    ampl.param["m"] = m
    ampl.param["K"] = K
    ampl.param["r"] = r
    ampl.param["Nacum"] = {a: b for a, b in enumerate(Nacum)}
    ampl.param["N"] = N
    ampl.param["d"] = {restr[i]: j for i, j in enumerate(d)}
    ampl.param["C"] = {
        (a, b): C[i][j]
        for i, a in enumerate(restr)
        for j, b in enumerate(activos)
    }
    ampl.param["xi"] = {
        (a, n): xi[i,j]
        for i, a in enumerate(activos)
        for j, n in enumerate(range(1,N+1))
    }
    ampl.param["rho"] = rho
    ampl.param["alfa"] = alfa
    ampl.param["eps"] = posibles_eps[i_eps]
    ampl.param["beta"] = {fuentes[i]: j for i, j in enumerate(beta)}

    ampl.set_option("solver", "highs") # Seleccionar el solver "highs"
    ampl.set_option("solver_msg", 0) # Silenciar el mensaje de retorno de AMPL
    ampl.solve() # Resolvemos el problema
    x_AMPL = ampl.get_variable("x") # Aquí el valor de "x" es convertido a un data frame de AMPL
    x_df = x_AMPL.get_values().to_pandas() # Aquí el valor de "x" es convertido a un data frame de Pandas
    x = x_df['x.val'] # Obtenemos el valor de "x" (indexada del 1 a m)
    tau_AMPL = ampl.get_variable("tau") # Aquí el valor de "tau" es convertido a un data frame de AMPL
    tau_df = tau_AMPL.get_values().to_pandas() # Aquí el valor de "x" es convertido a un data frame de Pandas
    tau = tau_df['tau.val'][0] # Obtenemos el valor de "tau"
    perdidas_pool[i_eps,prueba] = perdida(x, tau, m, rho, alfa) # Aproximamos la pérdida fuera de muestra
    certificados_pool[i_eps,prueba] = ampl.get_objective("coste").value() # Obtenemos el valor objetivo (el certificado)
    # Ajustamos la aproximación de la fiabilidad
    if certificados_pool[i_eps,prueba] >= perdidas_pool[i_eps,prueba]:
      fiabilidad_pool[i_eps] += 1

  # Utilizamos MoWB-DRO para cada combinación de radios
  for i_pareja in range(n_parejas_eps):
    # Problema MoWB-DRO
    ampl.eval(r"""
    reset;                          # Para limpiar cualquier modelo previo

    # Definimos los parámetros
    param m;                        # Número de activos financieros
    param K;                        # Número de fuentes de datos
    param r;                        # Número de restricciones del conjunto de incertidumbre
    set ACTIVOS := 1..m;            # Índices de los activos financieros
    set FUENTES := 1..K;            # Índices de las fuentes de datos
    set RESTR := 1..r;              # Índices de las restricciones del conjunto de incertidumbre

    set ACUM := 0..K;
    param Nacum {ACUM};             # Número acumulado de observaciones de cada fuente, comenzando por 0

    param N;                        # Número total de observaciones
    set OBS := 1..N;                # Índices de las observaciones
    param INDICES {OBS};            # Indica de qué fuente es cada observación (ordenadas por fuente)
    param xi {ACTIVOS, OBS};        # Conjunto de observaciones de los rendimientos anuales

    # Parámetros que definen el conjunto de incertidumbre
    param d {RESTR};
    param C {RESTR, ACTIVOS};

    param rho >= 0;                  # Codifica el riesgo
    param alfa >= 0;                 # Significación
    param eps {FUENTES} >= 0;        # Radio de las bolas de Wasserstein
    param beta {FUENTES} >= 0;       # Peso de cada fuente (basado en la confianza en este)

    # Definimos las variables de decisión
    var x {ACTIVOS} >= 0;            # Porcentaje a invertir en cada activo
    var tau;
    var lambda {FUENTES} >=0;
    var s {OBS};
    var y1 {RESTR, OBS}>=0;
    var y2 {RESTR, OBS}>=0;

    # Definimos el problema de optimización
    minimize coste:
        sum{k in FUENTES} beta[k] * (lambda[k] * eps[k] + (sum{i in (Nacum[k-1]+1)..Nacum[k]} s[i])/(Nacum[k] - Nacum[k-1]));
    s.t. prob_simplex:
          sum{a in ACTIVOS} x[a] = 1;
        restric1 {i in OBS}:
          rho*tau - (sum{a in ACTIVOS} x[a]*xi[a,i]) + sum{b in RESTR} y1[b,i]*(d[b]- (sum{a in ACTIVOS} C[b,a]*xi[a,i])) <= s[i];
        restric2 {i in OBS}:
          rho* (1 - 1/alfa) *tau - (1 + rho/alfa) * (sum{a in ACTIVOS} x[a]*xi[a,i]) + sum{b in RESTR} y2[b,i]*(d[b]- (sum{a in ACTIVOS} C[b,a]*xi[a,i])) <= s[i];
        restric3 {a in ACTIVOS, i in OBS}:
          (sum{b in RESTR} C[b,a]*y1[b,i]) + x[a] <= lambda[(INDICES[i])];
        restric4 {a in ACTIVOS, i in OBS}:
          (sum{b in RESTR} C[b,a]*y2[b,i]) + (1 + rho/alfa) * x[a] <= lambda[(INDICES[i])];
    """)

    # Definimos los datos en AMPL
    ampl.param["m"] = m
    ampl.param["K"] = K
    ampl.param["r"] = r
    ampl.param["Nacum"] = {a: b for a, b in enumerate(Nacum)}
    ampl.param["N"] = N
    ampl.param["INDICES"] = {a+1: b for a, b in enumerate(indices)}
    ampl.param["d"] = {restr[i]: j for i, j in enumerate(d)}
    ampl.param["C"] = {
        (a, b): C[i][j]
        for i, a in enumerate(restr)
        for j, b in enumerate(activos)
    }
    ampl.param["xi"] = {
        (a, n): xi[i,j]
        for i, a in enumerate(activos)
        for j, n in enumerate(range(1,N+1))
    }
    ampl.param["rho"] = rho
    ampl.param["alfa"] = alfa
    eps = [parejas_eps[i_pareja][0],parejas_eps[i_pareja][1]]
    ampl.param["eps"] = {fuentes[i]: j for i, j in enumerate(eps)}
    ampl.param["beta"] = {fuentes[i]: j for i, j in enumerate(beta)}

    ampl.set_option("solver", "highs") # Seleccionar el solver "highs"
    ampl.set_option("solver_msg", 0) # Silenciar el mensaje de retorno de AMPL
    ampl.solve() # Resolvemos el problema
    x_AMPL = ampl.get_variable("x") # Convertimos el valor de "x"  a un data frame de AMPL
    x_df = x_AMPL.get_values().to_pandas() # Convertimos el valor de "x" a un data frame de Pandas
    x = x_df['x.val'] # Obtenemos el valor de "x" (indexada del 1 a m)
    tau_AMPL = ampl.get_variable("tau") # Convertimos el valor de "tau" a un data frame de AMPL
    tau_df = tau_AMPL.get_values().to_pandas() # Convertimos el valor de "tau" a un data frame de Pandas
    tau = tau_df['tau.val'][0] # Obtenemos el valor de "tau"
    perdidas_mowb[i_pareja,prueba] = perdida(x, tau, m, rho, alfa) # Aproximamos la pérdida fuera de muestra
    certificados_mowb[i_pareja,prueba] = ampl.get_objective("coste").value() # Obtenemos el valor objetivo (el certificado)
    # Ajustamos la aproximación de la fiabilidad
    if certificados_mowb[i_pareja,prueba] >= perdidas_mowb[i_pareja,prueba]:
      fiabilidad_mowb[i_pareja] += 1

  # Problema SAA
  ampl.eval(r"""
  reset;                          # Para limpiar cualquier modelo previo

  # Definimos los parámetros
  param m;                        # Número de activos financieros
  set ACTIVOS := 1..m;            # Índices de los activos financieros
  param N;                        # Número total de observaciones
  set OBS := 1..N;                # Índices de las observaciones
  param xi {ACTIVOS, OBS};        # Conjunto de observaciones de los rendimientos anuales
  param rho >= 0;                 # Codifica el riesgo
  param alfa >= 0;                # Significación

  # Definimos las variables de decisión
  var x {ACTIVOS} >= 0;           # Porcentaje a invertir en cada activo
  var tau;
  var s {OBS};

  # Definimos el problema de optimización
  minimize coste:
      (sum{n in OBS} s[n])/N;
  s.t. prob_simplex:
        sum{a in ACTIVOS} x[a] = 1;
      restric1 {n in OBS}:
        rho*tau - (sum{a in ACTIVOS} x[a]*xi[a,n]) <= s[n];
      restric2 {n in OBS}:
        rho*(1 - 1/alfa)*tau - (1 + rho/alfa) * (sum{a in ACTIVOS} x[a]*xi[a,n]) <= s[n];
  """)

  # Definimos los datos en AMPL
  ampl.param["m"] = m
  ampl.param["N"] = N
  ampl.param["xi"] = {
      (a, n): xi[i,j]
      for i, a in enumerate(activos)
      for j, n in enumerate(range(1,N+1))
  }
  ampl.param["rho"] = rho
  ampl.param["alfa"] = alfa


  ampl.set_option("solver", "highs") # Seleccionar el solver "highs"
  ampl.set_option("solver_msg", 0) # Silenciar el mensaje de retorno de AMPL
  ampl.solve() # Resolvemos el problema
  x_AMPL = ampl.get_variable("x") # Convertimos el valor de "x" a un data frame de AMPL
  x_df = x_AMPL.get_values().to_pandas() # Convertimos el valor de "x" a un data frame de Pandas
  x = x_df['x.val'] # Obtenemos el valor de "x" (indexada del 1 a m)
  tau_AMPL = ampl.get_variable("tau") # Convertimos el valor de "tau" a un data frame de AMPL
  tau_df = tau_AMPL.get_values().to_pandas() # Convertimos el valor de "x" a un data frame de Pandas
  tau = tau_df['tau.val'][0] # Obtenemos el valor de "tau"
  perdidas_saa[prueba] = perdida(x, tau, m, rho, alfa) # Aproximamos la pérdida fuera de muestra
  certificados_saa[prueba] = ampl.get_objective("coste").value() # Obtenemos el valor objetivo (el certificado)
  # Ajustamos la aproximación de la fiabilidad
  if certificados_saa[prueba] >= perdidas_saa[prueba]:
    fiabilidad_saa += 1

# Imprimimos los resultados
for i_eps in range(n_eps):
  print("")
  print(f"Resultados de Pool-DRO con epsilon={posibles_eps[i_eps]}")
  print(f"Pérdida esperada: {np.mean(perdidas_pool[i_eps,:]):.3f}")
  print(f"Varianza de las pérdidas: {np.var(perdidas_pool[i_eps,:]):.3f}")
  print(f"Certificado esperado: {np.mean(certificados_pool[i_eps,:]):.3f}")
  print(f"Varianza de los certificados: {np.var(certificados_pool[i_eps,:]):.3f}")
  print(f"Fiabilidad: {fiabilidad_pool[i_eps]/repeticiones:.3f}")

for i_parejas in range(n_parejas_eps):
  print("")
  print(f"Resultados de MoWB-DRO con epsilon_1={parejas_eps[i_parejas][0]} y epsilon_2={parejas_eps[i_parejas][1]}")
  print(f"Pérdida esperada: {np.mean(perdidas_mowb[i_parejas,:]):.3f}")
  print(f"Varianza de las pérdidas: {np.var(perdidas_mowb[i_parejas,:]):.3f}")
  print(f"Certificado esperado: {np.mean(certificados_mowb[i_parejas,:]):.3f}")
  print(f"Varianza de los certificados: {np.var(certificados_mowb[i_parejas,:]):.3f}")
  print(f"Fiabilidad: {fiabilidad_mowb[i_parejas]/repeticiones:.3f}")

print("")
print(f"Resultados de SAA")
print(f"Pérdida esperada: {np.mean(perdidas_saa):.3f}")
print(f"Varianza de las pérdidas: {np.var(perdidas_saa):.3f}")
print(f"Certificado esperado: {np.mean(certificados_saa):.3f}")
print(f"Varianza de los certificados: {np.var(certificados_saa):.3f}")
print(f"Fiabilidad: {fiabilidad_saa/repeticiones:.3f}")
```

```python
# Definimos los parámetros
m=4 # Número de activos financieros
K=2 # Número de fuentes de datos
r=8 # Número de restricciones del conjunto del conjunto de incertidumbre
activos = range(1,m+1) # Índices de los activos financieros
fuentes = range(1,K+1) # Índices de las fuentes de datos
restr = range(1,r+1) # Índices de las restricciones del conjunto de incertidumbre

Ns = [30,30] # Número de observaciones de cada fuente
Nacum = [0] # Número acumulado de observaciones de cada fuente, comenzando por 0
for i in range(K):
  Nacum = Nacum + [Nacum[i]+Ns[i]]
N = sum(Ns) # Número de observaciones totales

indices = [] # Indica de que fuente es cada observacion (ordenadas por fuente)
for k in range(K):
  for i in range(Ns[k]):
    indices.append(k+1)

# Parámetros que definen el conjunto de incertidumbre
d = 10 * np.ones(r)
C = np.vstack((np.eye(m),-np.eye(m)))

beta = [1/2,1/2] # Pesos de cada fuente de datos
rho = 10 # Parámetro que codifica el riesgo que queremos asumir
alfa = 0.2 # Significación del Valor en Riesgo Condicional
eps = [0.01,0.01] # Radio de las bolas de Wasserstein

# Definimos las dos muestras
xi = np.zeros((K,m,Ns[0]+Ns[1]))
for i in range(Ns[0]):
  common = np.random.uniform(low=-2.0, high=2.0)
  for a in range(2):
    xi[0,a,i] = common + np.random.uniform(low=4.0, high=6.0)
  for a in range(2,m):
    xi[0,a,i] = common + np.random.uniform(low=-6.0, high=-4.0)
for i in range(Ns[1]):
  common = np.random.uniform(low=-1.0, high=2.0)
  for a in range(2):
    xi[1,a,i] = common + np.random.uniform(low=3.0, high=7.0)
  for a in range(2,m):
    xi[1,a,i] = common + np.random.uniform(low=-4.0, high=-2.0)

# Definimos la unión de las muestras
xijunto = np.zeros((m,Ns[0]+Ns[1]))
for i in range(Ns[0]):
  xijunto[0,i] = xi[0,0,i]
  xijunto[1,i] = xi[0,1,i]
for i in range(Ns[1]):
  xijunto[0,i+Ns[0]] = xi[1,0,i]
  xijunto[1,i+Ns[0]] = xi[1,1,i]

L = 500 # Número de iteraciones
h = [(1/(i+5))**(1/2) for i in range(L)] # Longitudes de paso

# Definimos los valores iniciales de las variables de decisión
xl = np.ones(m)*(1/m)
taul = 0

# Función que devuelve el subgradiente de l
def dl(x, tau, xi, m, rho, alfa):
  prod_esc = 0
  for i in range(m):
    prod_esc = prod_esc + x[i]*xi[i]
  l1 = -prod_esc + rho *tau
  l2 = - (1 + rho/alfa)*prod_esc + rho*(1 - 1/alfa)*tau
  l = max(l1,l2)
  if l == l1:
    sol = np.concatenate((-xi,np.array([rho])))
  else:
    sol = np.concatenate((-(1 + rho/alfa)*xi,np.array([rho*(1 - 1/alfa)])))
  return sol

# Realizamos las iteraciones
for l in range(L):

  fk = np.zeros(K) # Valor de fk(xl, taul) para cada cliente
  vk = np.zeros((K,m+1)) # Subgradientes pertinentes de cada cliente

  # Cálculos de cada cliente
  for k in range(K):

    # Calculamos la distribución que alcanza el supremo de fk(x_l,tau_l)
    ampl.eval(r"""
    reset;                          # Para limpiar cualquier modelo previo

    # Definimos los parámetros
    param m;                        # Número de activos financieros
    set ACTIVOS := 1..m;            # Índices de los activos financieros
    param r;                        # Número de restricciones del conjunto de incertidumbre
    set RESTR := 1..r;              # Índices de las restricciones del conjunto de incertidumbre
    param Nk;                       # Número total de observaciones
    set OBS := 1..Nk;               # Índices de las observaciones
    param xi {ACTIVOS, OBS};        # Conjunto de observaciones de los rendimientos anuales

    # Parámetros que definen el conjunto de incertidumbre
    param d {RESTR};
    param C {RESTR, ACTIVOS};

    param rho >= 0;                 # Codifica el riesgo
    param alfa >= 0;                # Significación
    param eps >= 0;                 # Radio de la bola de Wasserstein

    param x {ACTIVOS} >= 0;         # Porcentaje a invertir en cada activo
    param tau;

    # Definimos las variables de decisión
    var alfa1 {OBS} >= 0;
    var alfa2 {OBS} >= 0;
    var q1 {ACTIVOS, OBS};
    var q2 {ACTIVOS, OBS};

    # Definimos el problema de optimización
    maximize coste:
        (1/Nk) * sum{i in OBS} ( (alfa1[i]*rho*tau - (sum{a in ACTIVOS} x[a]*(alfa1[i]*xi[a,i]- q1[a,i]))) +
          (alfa2[i]*rho*(1 - 1/alfa)*tau - (1 + rho/alfa)*(sum{a in ACTIVOS} x[a]*(alfa2[i]*xi[a,i] - q2[a,i]))) );
    s.t. restric1:
          sum{i in OBS} ((sum{a in ACTIVOS} abs(q1[a,i])) + (sum{a in ACTIVOS} abs(q2[a,i]))) <= eps*Nk;
        restric2 {i in OBS}:
          alfa1[i] + alfa2[i] = 1;
        restric3 {i in OBS, b in RESTR}:
          alfa1[i]*((sum{a in ACTIVOS} C[b,a]*xi[a,i])- d[b]) <= (sum{a in ACTIVOS} C[b,a]*q1[a,i]);
        restric4 {i in OBS, b in RESTR}:
          alfa2[i]*((sum{a in ACTIVOS} C[b,a]*xi[a,i])- d[b]) <= (sum{a in ACTIVOS} C[b,a]*q2[a,i]);
    """)

    # Definimos los parámetros en AMPL
    ampl.param["m"] = m
    ampl.param["r"] = r
    ampl.param["Nk"] = Ns[k]
    ampl.param["d"] = {restr[i]: j for i, j in enumerate(d)}
    ampl.param["C"] = {
        (a, b): C[i][j]
        for i, a in enumerate(restr)
        for j, b in enumerate(activos)
    }
    ampl.param["xi"] = {
        (a, n): xi[k,i,j]
        for i, a in enumerate(activos)
        for j, n in enumerate(range(1,Ns[k]+1))
    }
    ampl.param["rho"] = rho
    ampl.param["alfa"] = alfa
    ampl.param["eps"] = eps[k]
    ampl.param["x"] = {a: xl[i] for i, a in enumerate(activos)}
    ampl.param["tau"] = taul

    ampl.set_option("solver", "highs") # Seleccionamos el solver "highs"
    ampl.set_option("solver_msg", 0) # Silenciamos el mensaje de retorno de AMPL
    ampl.solve() # Resolvemos el problema

    # Obtenemos el valor objetivo y lo almacenamos como el valor de fk(x_l,tau_l)
    fk[k] = ampl.get_objective("coste").value()

    alfa1_AMPL = ampl.get_variable("alfa1") # Convertimos el valor de "alfa1" a un data frame de AMPL.
    alfa1_df = alfa1_AMPL.get_values().to_pandas().round(4) # Convertimos el valor de "alfa1" a un data frame de Pandas
    alfa1 = alfa1_df['alfa1.val'] # Obtenemos el valor de "alfa1" (indexado de 1 a m)
    alfa2_AMPL = ampl.get_variable("alfa2") # Convertimos el valor de "alfa12" a un data frame de AMPL.
    alfa2_df = alfa2_AMPL.get_values().to_pandas().round(4) # Convertimos el valor de "alfa2" a un data frame de Pandas
    alfa2 = alfa2_df['alfa2.val'] # Obtenemos el valor de "alfa2" (indexado de 1 a m)
    q1_AMPL = ampl.get_variable("q1") # Convertimos el valor de "q1" a un data frame de AMPL.
    q1_df = q1_AMPL.get_values().to_pandas().round(4) # Convertimos el valor de "q1" a un data frame de Pandas
    q1 = q1_df['q1.val'] # Obtenemos el valor de "q1" (indexado de 1 a m)
    q2_AMPL = ampl.get_variable("q2") # Convertimos el valor de "q2" a un data frame de AMPL.
    q2_df = q2_AMPL.get_values().to_pandas().round(4) # Convertimos el valor de "q2" a un data frame de Pandas
    q2 = q2_df['q2.val'] # Obtenemos el valor de "q2" (indexado de 1 a m)

    # Calculamos los puntos que soportan la distribución que alcanza el supremo de fk(x_l,tau_l)
    xi_opt1 = xi[k,:,:]
    for i in range(Ns[k]):
      if alfa1[i+1] > 0:
        for j in range(m):
          xi_opt1[j,i] -= q1[j+1,i+1]/alfa1[i+1]
    xi_opt2 = xi[k,:,:]
    for i in range(Ns[k]):
      if alfa2[i+1] > 0:
        for j in range(m):
          xi_opt2[j,i] -= q2[j+1,i+1]/alfa2[i+1]

    # Calculamos el subgradiente del cliente K
    for i in range(Ns[k]):
      vk[k,:] += dl(xl, taul, xi_opt1[:,i], m, rho, alfa) * alfa1[i+1]
      vk[k,:] += dl(xl, taul, xi_opt2[:,i], m, rho, alfa) * alfa2[i+1]
    vk[k,:] = vk[k,:]/Ns[k]

  # Calculamos el valor objetivo
  fl = 0
  for k in range(K):
    fl += beta[k]*fk[k]

  # Actualizamos los mejores valores obtenidos hasta el momento
  if l==0:
    f_mejor = fl
    x_mejor = xl
  else:
    f_mejor = min(fl, f_mejor)
    if f_mejor==fl:
      x_mejor = xl

  # Calculamos el subgradiente de f
  vl = np.zeros(m+1)
  for k in range(K):
    vl += beta[k]*vk[k,:]

  # Calculamos el nuevo valor de x_l dando un paso y realizando la proyección euclidiana
  xl = xl - h[l]*vl[0:m]

  # Problema a resolver para hacer la proyección euclidiana
  ampl.eval(r"""
  reset;                          # Para limpiar cualquier modelo previo

  # Definimos los parámetros
  param m;                        # Número de activos financieros
  set ACTIVOS := 1..m;            # Índices de los activos financieros
  param x {ACTIVOS};              # Porcentaje a invertir en cada activo

  # Definimos las variables de decisión
  var xnew {ACTIVOS} >= 0;

  # Definimos el problema de optimización
  minimize distancia:
        sum{a in ACTIVOS} (xnew[a] - x[a])*(xnew[a] - x[a]);
  s.t. restric:
        sum{a in ACTIVOS} xnew[a] = 1;
  """)

  # Definimos los parámetros en AMPL
  ampl.param["m"] = m
  ampl.param["x"] = {a: xl[i] for i, a in enumerate(activos)}

  ampl.set_option("solver", "highs") # Seleccionamos el solver "highs"
  ampl.set_option("solver_msg", 0) # Silenciamos el mensaje de retorno de AMPL
  ampl.solve() # Resolvemos el problema
  xnew_AMPL = ampl.get_variable("xnew") # Convertimos el valor de "xnew" a un data frame de AMPL.
  xnew_df = xnew_AMPL.get_values().to_pandas().round(4) # Convertimos el valor de "xnew" a un data frame de Pandas
  xnew = xnew_df['xnew.val'] # Obtenemos el valor de "xnew" indexado de 1 a m

  # Actualizamos las variables de decisión
  for a in range(m):
    xl[a] = xnew[a+1]
  taul = taul - h[l]*vl[m]
print("Solución Algoritmo 1: ", x_mejor) # Mostramos la solución del Algoritmo 1


# Resolvemos ahora utilizando MoWB-DRO
ampl.eval(r"""
reset;                          # Para limpiar cualquier modelo previo

# Definimos los parámetros
param m;                        # Número de activos financieros
param K;                        # Número de fuentes de datos
param r;                        # Número de restricciones del conjunto de incertidumbre
set ACTIVOS := 1..m;            # Índices de los activos financieros
set FUENTES := 1..K;            # Índices de las fuentes de datos
set RESTR := 1..r;              # Índices de las restricciones del conjunto de incertidumbre

set ACUM := 0..K;
param Nacum {ACUM};             # Número acumulado de observaciones de cada fuente, comenzando por 0

param N;                        # Número total de observaciones
set OBS := 1..N;                # Índices de las observaciones
param INDICES {OBS};            # Indica de qué fuente es cada observación (ordenadas por fuente)
param xi {ACTIVOS, OBS};        # Conjunto de observaciones de los rendimientos anuales

# Parámetros que definen el conjunto de incertidumbre
param d {RESTR};
param C {RESTR, ACTIVOS};

param rho >= 0;                  # Codifica el riesgo
param alfa >= 0;                 # Significación
param eps {FUENTES} >= 0;        # Radio de las bolas de Wasserstein
param beta {FUENTES} >= 0;       # Peso de cada fuente (basado en la confianza en este)

# Definimos las variables de decisión
var x {ACTIVOS} >= 0;            # Porcentaje a invertir en cada activo
var tau;
var lambda {FUENTES} >=0;
var s {OBS};
var y1 {RESTR, OBS}>=0;
var y2 {RESTR, OBS}>=0;

# Definimos el problema de optimización
minimize coste:
    sum{k in FUENTES} beta[k] * (lambda[k] * eps[k] + (sum{i in (Nacum[k-1]+1)..Nacum[k]} s[i])/(Nacum[k] - Nacum[k-1]));
s.t. prob_simplex:
      sum{a in ACTIVOS} x[a] = 1;
    restric1 {i in OBS}:
      rho*tau - (sum{a in ACTIVOS} x[a]*xi[a,i]) + sum{b in RESTR} y1[b,i]*(d[b]- (sum{a in ACTIVOS} C[b,a]*xi[a,i])) <= s[i];
    restric2 {i in OBS}:
      rho* (1 - 1/alfa) *tau - (1 + rho/alfa) * (sum{a in ACTIVOS} x[a]*xi[a,i]) + sum{b in RESTR} y2[b,i]*(d[b]- (sum{a in ACTIVOS} C[b,a]*xi[a,i])) <= s[i];
    restric3 {a in ACTIVOS, i in OBS}:
      (sum{b in RESTR} C[b,a]*y1[b,i]) + x[a] <= lambda[(INDICES[i])];
    restric4 {a in ACTIVOS, i in OBS}:
      (sum{b in RESTR} C[b,a]*y2[b,i]) + (1 + rho/alfa) * x[a] <= lambda[(INDICES[i])];
""")

# Definimos los parámetros en AMPL
ampl.param["m"] = m
ampl.param["K"] = K
ampl.param["r"] = r
ampl.param["Nacum"] = {a: b for a, b in enumerate(Nacum)}
ampl.param["N"] = N
ampl.param["d"] = {restr[i]: j for i, j in enumerate(d)}
ampl.param["C"] = {
    (a, b): C[i][j]
    for i, a in enumerate(restr)
    for j, b in enumerate(activos)
}
ampl.param["xi"] = {
    (a, n): xijunto[i,j]
    for i, a in enumerate(activos)
    for j, n in enumerate(range(1,N+1))
}
ampl.param["rho"] = rho
ampl.param["alfa"] = alfa
ampl.param["INDICES"] = {a+1: b for a, b in enumerate(indices)}
ampl.param["eps"] = {fuentes[i]: j for i, j in enumerate(eps)}
ampl.param["beta"] = {fuentes[i]: j for i, j in enumerate(beta)}

ampl.set_option("solver", "highs") # Seleccionamos el solver "highs"
ampl.set_option("solver_msg", 0) # Silenciamos el mensaje de retorno de AMPL
ampl.solve() # Resolvemos el problema
x_AMPL = ampl.get_variable("x") # Convertimos el valor de "x" a un data frame de AMPL.
x_df = x_AMPL.get_values().to_pandas().round(4) # Convertimos el valor de "x" a un data frame de Pandas
x = x_df['x.val'] # Obtenemos el valor de "x" indexado de 1 a m
x_sol = np.zeros(m) # Valor de la solución de MoWB-DRO indexada de 0 a (m-1)
for a in range(m):
  x_sol[a] = x[a+1]
print("Solución MoWB-DRO: ", x_sol) # Mostramos la solución de MoWB-DRO
```
